### Cleaning Customer Transactions Table

Capitalize last names properly

```python
ct
```

```python
ct.last_name = ct.last_name.str.capitalize()
ct
```

Capitalize first names properly

```python
ct.first_name = ct.first_name.str.capitalize()
ct
```

Remove unnecesssary characters in names

```python
def filter(x):
    mat = re.search('[a-zA-Z]+', x)
    return mat.group() if mat else mat
```

```python
ct.last_name
```

```python
ct.last_name = ct.last_name.apply(filter)
ct
```

```python
ct.last_name[ct.last_name != ct.last_name.apply(filter)]
```

```python
ct.first_name
```

```python
ct.first_name = ct.first_name.apply(filter)
ct
```

```python
ct.first_name[ct.first_name != ct.first_name.apply(filter)]
```

Convert date string to datetime values

```python
ct.birthday
```

```python
ct.birthday = pd.to_datetime(ct.birthday)
```
